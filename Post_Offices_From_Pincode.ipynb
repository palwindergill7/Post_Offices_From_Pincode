# Get Post Offices data from Pincode

```python
import requests
import json
import pandas as pd
pincode= pd.read_csv('Pincode.csv')
#To Create DataFrame
pinlist= ['142050']
d=requests.get('https://api.postalpincode.in/pincode/142050')
data=d.json()
#create data frame for Pincode Storage
PinData= pd.DataFrame(columns=data[0]['PostOffice'][0].keys())
```

##### Enter the Pincodes here
Use "," Commas to enter multiple Pincodes

```python
PINCODES= input(' Enter the Pincodes here: ')
pinlist = PINCODES.split(',')

#pindata= pd.read_csv('PINCODE.csv')
#pinlist= list(pindata['Pincode'])
```

```python
#for pin in pinlist[1:10]:
for pin in pinlist:
    url="https://api.postalpincode.in/pincode/"+str(pin)
    d=requests.get(url)
    data=d.json()
    
    
    if(data[0]['Status']=='Success'):
        for i in range(0,len(data[0]['PostOffice'])):
    
            x=(data[0]['PostOffice'][i].values())
            PinData.loc[len(PinData)]=list(x)
    else:
        PinData.loc[len(PinData)]=pin
  
```
